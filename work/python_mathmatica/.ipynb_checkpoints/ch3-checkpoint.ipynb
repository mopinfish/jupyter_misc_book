```python
# 体重の折れ線グラフを描画する

%matplotlib inline
import matplotlib.pyplot as plt

## データセット
x = [1, 2, 3, 4, 5, 6, 7]
y = [64.8, 64.0, 68.8, 56.4, 60.3, 60.2, 68.9]

## グラフを描画
plt.plot(x, y)
plt.grid(color = '0.8')
plt.show()
```

```python
# 一次関数 y = 3 * x - 24 を描画
%matplotlib inline
import matplotlib.pyplot as plt

# データセット
x = list(range(1, 11)) # xの範囲（１〜１０）
y = []
for i in range(10):
    y.append(3 * i - 24) # y = 3x - 24

# グラフを描画
plt.plot(x, y)
plt.grid(color = '0.8')
plt.show()
```

```python
# 一次関数 y = 3 * x - 24 を描画
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

# データセット
x = np.array(range(1, 11)) # xの範囲（１〜１０）
y = 3 * x - 24

# グラフを描画
plt.plot(x, y)
plt.grid(color = '0.8')
plt.show()
```

```python
# 連立方程式をとく
from sympy import Symbol, solve


# 式を定義
a = Symbol('a')
b = Symbol('b')
ex1 = a + b - 5
ex2 = 2 * a + b - 8

# 連立方程式の解を求める
solve((ex1, ex2))
```

```python
## 直行する軸と交点

%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(-1, 6) # x軸の値
y = 1/2 * x - 1/2
y2 = -2 * x + 7

# グラフを描画
plt.plot(x, y)
plt.plot(x, y2)
plt.axis('equal')
plt.grid(color = '0.8')
plt.show()
```

```python
## 直線の交点を求める
from sympy import Symbol, solve
# 文字の定義
x = Symbol('x')
y = Symbol('y')
ex1 = -3/2 * x + 6 - y
ex2 = 1/2 * x + 2 - y
print(solve((ex1, ex2)))

# グラフで確認
ex = np.arange(-1, 6)
y1 = -3/2 * ex + 6
y2 = 1/2 * ex + 2
plt.plot(ex, y1)
plt.plot(ex, y2)
plt.axis('equal')
plt.grid(color = '0.8')
plt.show()
```

```python
# 垂直二等分線を求める

# 元の式
a1 = (5 - 1) / (6 - 0)
b1 = 1

# 線分の中点
cx = (6 + 0) / 2
cy = (5 + 1) / 2

# 線分に直交する式の傾き
a2 = -1 / a1
# 線分に直交する式の切片(b2 = y - a2 * x)
b2 = cy - a2 * cx

# 直線の式
x = np.arange(0, 7) # xの範囲
y1 = a1 * x + b1 # 元の直線
y2 = a2 * x + b2 # 垂直二等分線

# 描画
plt.figure(figsize = (20, 6))
plt.plot(x, y1)
plt.plot(x, y2)

plt.xlabel('X')
plt.ylabel('Y')
plt.axis('equal')
plt.grid(color = '0.8')
plt.show()

```
