```python
import networkx as net
import random as r
```

```python
class Person(object):
    def __init__(self, id):
        # 初期状態の意見を一つ持つところからスタートする
        self.id = id
        self.i = r.random()
        self.a = self.i
        # 最初の見解もその後の情報も同じように評価する
        self.alpha = 0.8
    def __str__(self):
        return (str(self.id))
    
```

```python
density = 0.9
g = net.Graph()

## Personオブジェクトのネットワークを作る
for i in range(10):
    p = Person(i)
    g.add_node(p)

## こうすると、単純なランダムグラフが作られ、全てのノードペアは同じ確率で接続される
for x in g.nodes():
    for y in g.nodes():
        if r.random() <= density: g.add_edge(x, y)

## 得られたグラフを描画し、値によってノードに色を塗る
col = [n.a for n in g.nodes()]
pos = net.spring_layout(g)
net.draw_networkx(g, pos = pos, node_color = col)
```
