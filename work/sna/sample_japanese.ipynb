```python
import matplotlib.pyplot as plt
import networkx as nx
from IPython.display import SVG, display
```

```python
def draw(G):
    nx.draw_networkx(G)
    plt.show()
```

```python
G = nx.Graph()
# 頂点の追加
G.add_node('日本語')                
G.add_nodes_from([3, 4, 5])
draw(G)
```

```python
from matplotlib import pyplot as plt
plt.style.use('ggplot')
 
plt.figure(figsize=(12,5))
plt.plot(0,0)
plt.title("日本語")
```
